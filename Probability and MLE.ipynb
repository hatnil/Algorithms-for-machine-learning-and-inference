## Bayes Rules
# **Solution:**
We have the two incidents
\begin{align*}
    A &= \text{positive test} \\
    B &= \text{having disease}
\end{align*}
and are looking for $P(B|A)$ and can compute it using Bayes rule
\begin{align*}
    P(B|A) &= \frac{P(A|B)P(B)}{P(A)}
\end{align*}
where
\begin{align*}
    P(A) &= \sum_j P(A|B_j)P(B_j) \\
    &= P(A|B)P(B)+P(A|\overline{B})P(\overline{B})    
\end{align*}
such that we get
\begin{align}
    P(B|A) &= \frac{P(A|B)P(B)}{P(A|B)P(B)+P(A|\overline{B})P(\overline{B})}    
\end{align}
The text gives us
\begin{align}
    P(A|B) &= 0.992 \\
    P(\overline{A}|B) &= 1-P(A|B) = 0.008 \\
    P(\overline{A}|\overline{B}) &= 0.992 \\
    P(A|\overline{B}) &= 1-P(\overline{A}|\overline{B}) = 0.008 \\
    P(B) &= \frac{1}{1000} = 0.001 \\
    P(\overline{B}) &= 1-P(B)=0.999
\end{align}
Inserting the values from the text into our problem equation gives us
\begin{align}
    P(B|A) &= \frac{0.992\cdot0.001}{0.992\cdot0.001+0.008\cdot0.999} = 0.1104 \approx 11\%
\end{align}
So actually being ill has a probability of 11\%.  \\
Changing the value of $P(A|B)$ and $P(\overline{A}|\overline{B})$ to $0.92$ gives us
\begin{align}
    P(B|A) &= \frac{0.92\cdot0.001}{0.92\cdot0.001+0.08\cdot0.999} = 0.0114 \approx 1\%
\end{align}

## **Setting hyperparameters**

We have the probability density function $f$ of the beta distribution $B(\theta)$ with the following properties
\begin{align}
    f(x;\theta) &= f(x;a^2,b) = \frac{1}{B(a^2,b)}x^{a^2-1}(1-x)^{b-1}
\end{align}
where $\{a^2,b\} \in \theta$. Some further properties for the later calculation
\begin{align}
    B(a^2,b) &= \int_0^1 x^{a^2-1}(1-x)^{b-1}dx= \frac{\Gamma(a^2)\Gamma(b)}{\Gamma(a^2+b)} \\
    \Gamma(a^2 + 1) &= a^2 \Gamma(a^2) \\
    \Gamma(a^2 + 2) &= a^2 (a^2+1) \Gamma(a^2)
\end{align}
So we get the mean 
\begin{align}
    m &= E[\theta] \\
    &= \int_0^1 x\frac{1}{B(a^2,b)}x^{a^2-1}(1-x)^{b-1}dx \\
    &= \frac{\int_0^1 x^{a^2}(1-x)^{b-1}}{B(a^2,b)}dx \\
    &= \frac{B(a^2+1,b)}{B(a^2,b)}  \\
    &= \frac{\Gamma(a^2+1)\Gamma(b)}{\Gamma(a^2+1+b)} \frac{\Gamma(a^2+b)}{\Gamma(a^2)\Gamma(b)} \\
    &= \frac{a^2\Gamma(a^2)}{(a^2+b)\Gamma(a^2+b)} \frac{\Gamma(a^2+b)}{\Gamma(a^2)} \\
    m &= \frac{a^2}{a^2+b}
\end{align}
and the variance
\begin{align}
    v &= var(\theta) \\
    &= E\left[(\theta-m)^2\right]  \\
    &= E\left[\theta^2\right] -E\left[\theta\right]^2  \\
    &= E\left[\theta^2\right] -m^2\\
    &= E\left[\theta^2\right] -\left(\frac{a^2}{a^2+b}\right)^2
\end{align}
with
\begin{align}
    E\left[\theta^2\right]&= \int_0^1 x^2\frac{1}{B(a^2,b)}x^{a^2-1}(1-x)^{b-1}dx \\
    &= \frac{\int_0^1 x^{a^2+1}(1-x)^{b-1}}{B(a^2,b)}dx \\
    &= \frac{B(a^2+2,b)}{B(a^2,b)}  \\
    &= \frac{\Gamma(a^2+2)\Gamma(b)}{\Gamma(a^2+2+b)} \frac{\Gamma(a^2+b)}{\Gamma(a^2)\Gamma(b)} \\
    &= \frac{a^2(a^2+1)}{(a^2+b)(a^2+b+1)}
\end{align}
such that
\begin{align}
    v &= \frac{a^2(a^2+1)}{(a^2+b)(a^2+b+1)} -\left(\frac{a^2}{a^2+b}\right)^2 \\
    &= \frac{a^2(a^2+1)(a^2+b) - a^2(a^2+b+1)}{(a^2+b)^2(a^2+b+1)}\\
    &= \frac{a^6+ a^4 +a^4b +a^2b - a^6 -a^4b -a^4}{(a^2+b)^2(a^2+b+1)} \\
    &= \frac{a^2b }{(a^2+b)^2(a^2+b+1)}
\end{align}
Using the previous equations for m we get
\begin{align}
b &= \frac{a^2}{m}-a^2 \\
a^2+b &= \frac{a^2}{m} \\
\end{align}
inserting them in the variance yields
\begin{align}
v &= \frac{a^2(\frac{a^2}{m}-a^2)}{(\frac{a^2}{m})^2(\frac{a^2}{m}+1)} \\
v &= \frac{m-m^2}{(\frac{a^2}{m}+1)} \\
\frac{a^2}{m}+1 &= \frac{m-m^2}{v} \\
a^2 &= \frac{m^2-m^3}{v}-m \\
a &= \sqrt{\frac{m^2-m^3-vm}{v}} \\
\end{align}
We use this solution to insert it into the equation for b
\begin{align}
b &= \frac{a^2}{m}-a^2 = a^2(\frac{1}{m}-1)\\
b &= \frac{m^2-m^3-vm}{v}(\frac{1}{m}-1)
\end{align}

## **Correlation**
We make use of the linearity property of expectations to simplify the equation for the covariance.
\begin{align}
cov(X,Y) &= E\left[\left(X-E\left[X\right]\right)\left(Y-E\left[Y\right]\right)\right] \\
    &= E\left[XY-X E\left[Y\right]-E\left[X\right]Y +E\left[X\right] E\left[Y\right]\right] \\
    &= E\left[XY\right]-E\left[X\right]E\left[Y\right]-E\left[X\right] E\left[Y\right] +E\left[X\right]  E\left[Y\right] \\
    &= E\left[XY\right]-E\left[X\right]E\left[Y\right]
\end{align}
By inserting our given random variables $X$ and $Y$ we get the following
\begin{align}
    cov(X,Y) &= cov\left(X,X^4\right) \\
        &= E\left[X\cdot X^4\right]-E\left[X\right]E\left[X^4\right] \\
        &= E\left[X^5\right]-E\left[X\right]E\left[X^4\right] \\
        &= 0 - 0\cdot E\left[X^4\right] \\
        &= 0
\end{align}
In the following it is explained why the expected values $E\left[X\right]$ and $E\left[X^5\right]$ equal zero. The density function of the uniform distribution $g(x)$ is a constant function with a value depending on the width of the interval.
\begin{align}
    g(x) = 
    \begin{cases}
    \frac{1}{b-a} & \text{for } a\leq x \leq b\\
    0              & \text{otherwise}
    \end{cases}
\end{align}
So we can compute our expected values
\begin{align}
E\left[X\right]  &= \int_{-2}^2 x \cdot g(x) \, dx\\
    &= \int_{-2}^2 x \cdot \frac{1}{2-(-2)} \, dx\\
    &= \frac{1}{4}\left[\frac{1}{2}x^2\right]_{-2}^2 \\
    &= \frac{1}{8} (2^2- (-2)^2) \\
    &= (0.5 - 0.5) \\
    &= 0 \\
\end{align}
and
\begin{align}
    E\left[X^5\right]  &= \int_{-2}^2 x^5 \cdot g(x) \, dx\\
    &= \int_{-2}^2 x^5 \cdot \frac{1}{2-(-2)} \, dx\\
    &= \frac{1}{4}\left[\frac{1}{6}x^6\right]_{-2}^2 \\
    &= \frac{1}{24} (2^6- (-2)^6) \\
    &= \frac{1}{24} (64 - 64) \\
    &= 0 \\
\end{align}

```python
# Sample Answer to (b)
import numpy as np # numpy


def sge(X):
    """
    This function computes  the mean mu and standard deviation sigma 
    for a dataset coming from spherical Gaussian distribution N(mu, 2 sigma^2 I)

    @param X : A numpy array of size n x p where each row represents a p-dimensional data point
    e.g. X = [ [2, 1], [3, 7], [4 ,5] ] is a dataset with 3 samples having two coordinates each.
    @return mu: A numpy array of size p representing the estimated mean of the dataset
    @return sigma: A real number indicating the estimated standard deviation of the dataset
    """
    N=X.shape[1]
    M=X.shape[0]
    mu = np.zeros(X.shape[1])
    for n in range(N):
      sum=0
      for m in range(M):
        sum=sum+X[m][n]
      mu[n]=sum/M
      
    

    sigma = 0
    for i in X:
      product=((i-mu).T).dot(i-mu)
      sigma=sigma+product
    sigma=np.sqrt(sigma/(2*N*M))
      
    return mu, sigma
```
